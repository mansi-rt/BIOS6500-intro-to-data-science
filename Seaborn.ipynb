```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
#The majority of code examples in the documentation use the `load_dataset()` function to quickly access sample datasets. 
# These datasets aren't unique or special; they are simply pandas dataframes, and we could just as easily load them using `pandas.read_csv()` or manually construct them. 
# Seaborn is highly flexible and can work with a variety of data structures.

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

The sns.relplot function creates a relational plot, which is useful for exploring relationships between multiple variables in your dataset. Here’s a breakdown of each argument:

- data=tips: Specifies that the data to be used comes from the tips dataset.
- x="total_bill": Maps the total_bill column of the dataset to the x-axis.
- y="tip": Maps the tip column to the y-axis.
- col="time": Facets the plot into two columns, one for each value of the time variable (Lunch and Dinner).
- hue="smoker": Adds color encoding to differentiate between smokers and non-smokers.
- style="smoker": Adds different marker styles to visually distinguish between smokers and non-smokers (e.g., circles for non-smokers and X markers for smokers).
- size="size": Adjusts the size of the markers according to the size column, which represents the number of people in the dining party.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Create a FacetGrid to show all four datasets in different subplots
sns.set(style="ticks", palette="muted")

# Define the figure
g = sns.FacetGrid(anscombe, col="dataset", col_wrap=2, height=4)

# Plot scatter and linear regression lines for each dataset
g.map(sns.scatterplot, "x", "y", color="blue")
g.map(sns.lineplot, "x", "y", color="red", ci=None)

# Set titles and adjust the layout
g.set_titles("Dataset {col_name}")
g.set_axis_labels("X Values", "Y Values")
plt.tight_layout()
plt.show()
```

- sns.set(style="ticks", palette="muted"): This applies Seaborn’s default aesthetic settings for the plot, making it look more polished. The "muted" palette is used for soft, understated colors.

- FacetGrid: The FacetGrid is a powerful function in Seaborn used to create a grid of subplots. The col="dataset" argument means each subplot will represent one of the four datasets (I, II, III, IV).

- col_wrap=2 creates a 2x2 grid.
- height=4 sets the height of each subplot.
- g.map(sns.scatterplot, "x", "y", color="blue"): This function maps the x and y values to scatter plots for each dataset (I-IV) and uses blue points.

- g.map(sns.lineplot, "x", "y", color="red", ci=None): After plotting the scatter points, this line adds a linear regression line to each subplot using a red color. The ci=None argument - means no confidence interval is shown around the regression line.

- g.set_titles("Dataset {col_name}"): This sets the title of each subplot as the name of the dataset (I, II, III, or IV).

- g.set_axis_labels("X Values", "Y Values"): This labels the X and Y axes across all subplots.

- plt.tight_layout(): This ensures that the subplots don’t overlap or have excessive padding, making the layout look cleaner.

```python
import pandas as pd

# Calculate basic summary statistics for each dataset
summary_stats = anscombe.groupby('dataset').agg(
    mean_x=('x', 'mean'),
    mean_y=('y', 'mean'),
    variance_x=('x', 'var'),
    variance_y=('y', 'var'),
    correlation=('x', lambda x: x.corr(anscombe.loc[anscombe['dataset'] == x.name, 'y']))
)

# Display the summary statistics
print(summary_stats)
```

- anscombe.groupby('dataset'): This groups the dataset by the dataset column, which corresponds to the four datasets (I, II, III, and IV). This ensures that calculations are done separately for each dataset.

- .agg(): The agg() function is used to apply multiple aggregation operations on different columns.
- x.corr(): This computes the Pearson correlation between the x values and the corresponding y values for each dataset.

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

In fMRI research, the relationships between brain regions and events (stimuli) are crucial for understanding brain function. Different regions of the brain are specialized for different tasks, and the timing of their activation (represented by timepoint) can tell us a lot about how the brain processes stimuli.

- The frontal lobe is often associated with higher-order functions, such as planning and decision-making, and might be more active during tasks requiring attention or motor control ("stim" events).
- The parietal lobe is involved in processing sensory information and spatial orientation, and it may be more active during tasks involving sensory input ("cue" events).

By adjusting the plotting technique (changing hue, style, col, and row), we can explore the relationship between brain regions and stimuli from different perspectives, which is key for both exploratory data analysis and hypothesis-driven research in neuroscience.

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", row="event",
    hue="region", style="region",
)
```

Goal: The focus has shifted to comparing how different brain regions (e.g., frontal, parietal) behave during each event (stim or cue) over time.
- row="event": The subplots are now organized by event, with separate rows for each event (e.g., "stim" and "cue"). This means that each row shows the time series of signals for one particular stimulus event.
- hue="region": The different brain regions are colored differently, allowing us to see how each region responds during each event.
- style="region": The different line styles now correspond to brain regions, helping to further distinguish them.

Additional comments:
- By faceting the plot with row="event", we are focusing on how brain activity behaves during each event. Instead of splitting by region (like in the original plot), we are now asking: How do different brain regions (frontal, parietal) respond during each specific type of stimulus (stim or cue)?

- In this case, we can compare the regions more directly within each event. For instance, during a "stim" event, we can see how different regions like the frontal or parietal cortex are activated.

- If, for example, the frontal region shows more significant activation during the "stim" event and the parietal region shows more during the "cue" event, this would align with known functional roles of these brain areas (the frontal lobe being involved in decision-making and motor control, while the parietal lobe is involved in sensory processing).

```python
import time
from tqdm import tqdm

# Simulate a long-running function
# This would help you see the progress of a function
def long_running_function():
    print("Starting the process...")
    for i in tqdm(range(5), desc="Processing", ncols=100):
        time.sleep(1)  # Simulating a process that takes time
    print("Process complete!")

# Execute the function
long_running_function()
```

```python
import sys
import time

# Custom loading animation
def loading_animation(duration):
    for _ in range(duration):
        for frame in ["-", "\\", "|", "/"]:
            sys.stdout.write(f"\rLoading... {frame}")
            sys.stdout.flush()
            time.sleep(0.25)
# Simulate a long-running function with a loading animation
def long_running_function():
    print("Starting the process...")
    loading_animation(20)  # 20 animation frames (~5 seconds)
    print("\nProcess complete!")

# Execute the function
long_running_function()
```
