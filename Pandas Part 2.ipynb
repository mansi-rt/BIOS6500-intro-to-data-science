```python
weather = pd.read_csv ('./data/weather.csv')
```

```python
weather
```

```python
weather_melt = pd.melt(
                weather,
                id_vars=['id','year','month','element'],
                var_name='day',
                 value_name ='temp'
)
```

```python
weather_melt
```

```python
weather_clean = weather_melt.pivot_table(
                        index=['id','year','month','day'],
                        columns='element',values='temp'
)
```

```python
weather_clean
```

```python
weather_clean.reset_index()
```
