## Bar color

This is an example showing how to control bar color and legend entries
using the *color* and *label* parameters of `~matplotlib.pyplot.bar`.
Note that labels with a preceding underscore won't show up in the legend.

plt.subplots() creates a figure (fig) and a set of axes (ax). The figure is the overall container that holds the plot, while the axes is the specific area in the figure where the plot is drawn.
ax will be used to manipulate the plot and set different attributes.


- fruits contains the labels (categories) for the x-axis.
- counts contains the corresponding values (height of the bars) for the y-axis.
- bar_labels contains the labels that will appear in the legend for each bar.
- bar_colors contains the colors used for each bar. These colors are specified using the tab color names, which are predefined color schemes in matplotlib (e.g., tab:red, tab:blue).

Note that bar_labels contains an entry '_red', where the leading underscore is used to exclude this entry from being shown in the legend.

- This line creates a bar chart using the ax.bar() method:
- fruits represents the categories on the x-axis (the labels).
- counts represents the heights of the bars, corresponding to each fruit.
- label=bar_labels specifies the labels that will be used in the legend for each bar.
- color=bar_colors specifies the colors of the bars.
- Each bar will be displayed in the color defined in bar_colors and associated with a corresponding label from bar_labels.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 79, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```

- ax.set_ylabel('fruit supply'): This sets the label for the y-axis, which represents the quantity of fruits.
- ax.set_title('Fruit supply by kind and color'): This sets the title of the plot, displayed at the top of the chart.
- ax.legend(title='Fruit color'): This adds a legend to the plot, where the title='Fruit color' gives the legend a title. The legend corresponds to the colors and labels specified in bar_labels. Since one of the labels starts with an underscore ('_red'), it is ignored and won't appear in the legend.

```python
import matplotlib.pyplot as plt
import numpy as np

# data from https://allisonhorst.github.io/palmerpenguins/

species = (
    "Adelie\n $\\mu=$3700.66g",
    "Chinstrap\n $\\mu=$3733.09g",
    "Gentoo\n $\\mu=5076.02g$",
)
weight_counts = {
    "Below": np.array([70, 31, 58]),
    "Above": np.array([82, 37, 66]),
}
width = 0.2

fig, ax = plt.subplots()
bottom = np.zeros(3)

for boolean, weight_count in weight_counts.items():
    p = ax.bar(species, weight_count, width, label=boolean, bottom=bottom)
    bottom += weight_count

ax.set_title("Number of penguins with above average body mass")
ax.legend(loc="upper right")

plt.show()
```

species: This tuple contains the names of the three penguin species.
penguin_means: This dictionary holds the three attributes ("Bill Depth," "Bill Length," and "Flipper Length") as keys, with their corresponding mean measurements (in millimeters) for each species as values in tuples. For example:
Bill Depth is 18.35mm for Adelie, 18.43mm for Chinstrap, and 14.98mm for Gentoo.
Similarly, mean measurements for Bill Length and Flipper Length are stored.

x = np.arange(len(species)): This creates an array with the positions for the bars on the x-axis. Since there are 3 species, x will be an array: [0, 1, 2], representing the positions for Adelie, Chinstrap, and Gentoo.
width = 0.25: Specifies the width of each bar. This ensures that the bars representing different attributes are placed close to each other in groups.
multiplier: This variable helps shift the bars horizontally so they don't overlap and are grouped side by side.

```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

This line creates a figure (fig) and axes (ax) object using plt.subplots(). The layout='constrained' argument ensures the layout is automatically adjusted to prevent overlapping elements (like labels and titles).

This for loop iterates over each attribute (like "Bill Depth") and its corresponding mean measurements.
For each attribute, it:
Calculates the offset as width * multiplier to shift each set of bars horizontally, ensuring that the bars for different attributes dont overlap.
Creates a set of bars using ax.bar() with the x + offset for proper positioning. measurement represents the height of each bar, and width defines the width of each bar.
Adds labels to the bars with ax.bar_label(rects, padding=3), which displays the numeric values (heights) directly above each bar.
Increments the multiplier to move the next set of bars further to the right.

ax.set_ylabel('Length (mm)'): Sets the label for the y-axis, indicating that the measurements are in millimeters.
ax.set_title('Penguin attributes by species'): Sets the title of the plot.
ax.set_xticks(x + width, species): Customizes the x-axis ticks to be centered below the groups of bars, with the names of the species as labels.
ax.legend(loc='upper left', ncols=3): Adds a legend to the plot, showing which color corresponds to which attribute (e.g., Bill Depth, Bill Length, Flipper Length). The ncols=3 arranges the legend items in three columns, and loc='upper left' places the legend in the upper-left corner.
ax.set_ylim(0, 250): Sets the y-axis limit from 0 to 250 to ensure all bars fit within the chart.

```python
# Fixing random state for reproducibility
np.random.seed(19880831)

fig, ax = plt.subplots()

# Example data
people = ('Alice', 'Bob', 'Charlie', 'Debbie', 'Eva')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

This line creates a figure (fig) and an axes (ax). The figure is the entire plotting area, while the axes is the specific area where the plot will be drawn.

ax.broken_barh() is used to create a horizontal bar plot with breaks.
- The first argument [(110, 30), (150, 10)] defines the intervals for the bar segments. Each tuple represents a segment, where the first value is the starting position and the second value is the length of the bar. In this case:
-   The first segment starts at 110 and has a width of 30.
-    The second segment starts at 150 and has a width of 10.
-The second argument (10, 9) defines the position and height of the bar on the y-axis. The bar is centered at y=10 and has a height of 9.
- facecolors='tab:blue' specifies the color of the bar (using one of matplotlib's built-in color schemes).

```python
# Horizontal bar plot with gaps
fig, ax = plt.subplots()
ax.broken_barh([(110, 30), (150, 10)], (10, 9), facecolors='tab:blue')
ax.broken_barh([(10, 50), (100, 20), (130, 10)], (20, 9),
               facecolors=('tab:orange', 'tab:green', 'tab:red'))
ax.set_ylim(5, 35)
ax.set_xlim(0, 200)
ax.set_xlabel('seconds since start')
ax.set_yticks([15, 25], labels=['Bill', 'Jim'])     # Modify y-axis tick labels
ax.grid(True)                                       # Make grid lines visible
ax.annotate('race interrupted', (61, 25),
            xytext=(0.8, 0.9), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=16,
            horizontalalignment='right', verticalalignment='top')

plt.show()
```

This line creates another broken bar with multiple segments:
- The first argument [(10, 50), (100, 20), (130, 10)] defines three segments:
- The first starts at 10 with a length of 50.
- The second starts at 100 with a length of 20.
- The third starts at 130 with a length of 10.
- The second argument (20, 9) places the bar centered at y=20 with a height of 9.
- facecolors=('tab:orange', 'tab:green', 'tab:red') specifies the colors for the three segments. The first segment is orange, the second is green, and the third is red.

- ax.set_ylim(5, 35) sets the y-axis limits, so the plot displays the range from 5 to 35 on the y-axis.
- ax.set_xlim(0, 200) sets the x-axis limits, so the plot displays the range from 0 to 200 on the x-axis.

- ax.set_xlabel('seconds since start') adds a label to the x-axis, indicating that the x-values represent seconds since the start.
- ax.set_yticks([15, 25], labels=['Bill', 'Jim']) sets the y-axis ticks at positions 15 and 25, and labels them as 'Bill' and 'Jim' respectively. This makes it clear which individual each horizontal bar corresponds to.

This enables grid lines, making it easier to visually align the bars with specific values on the axes.


- ax.annotate() is used to add an annotation to the plot.
- 'race interrupted' is the text of the annotation.
- (61, 25) defines the point (x=61, y=25) where the annotation is pointing (near the orange bar).
- xytext=(0.8, 0.9) places the annotation text at 80% of the width and 90% of the height of the axes (relative to the entire plot area).
- arrowprops=dict(facecolor='black', shrink=0.05) draws an arrow from the text to the point (61, 25), with a black face color and a small shrinking effect for the arrow.
- fontsize=16 sets the font size of the annotation text.
- horizontalalignment='right' and verticalalignment='top' align the annotation text to the right and top of its position.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```

- t: This line creates an array of time values from 0.0 to 2.0 (exclusive) with steps of 0.01. The result is an array of evenly spaced values that represent time (in seconds).
- s: This computes the sine wave values based on the time values in t. Specifically:
- np.sin(2 * np.pi * t) computes the sine of 2t, generating a periodic waveform with a period of 1 second.
- Adding 1 to the sine values shifts the entire wave vertically, so the wave oscillates between 0 and 2 instead of between -1 and 1.

- fig, ax = plt.subplots()
 This line creates a figure (fig) and axes (ax). The figure is the overall container for the plot, while the axes is the area where the plot is drawn.

- xlabel='time (s)': Sets the label for the x-axis, indicating that the values represent time in seconds.
- ylabel='voltage (mV)': Sets the label for the y-axis, indicating that the values represent voltage in millivolts (mV).
- title='About as simple as it gets, folks': Adds a title to the plot, providing a simple description of the plot.

- fig.savefig("test.png")
 This saves the plot as an image file named "test.png" in the current working directory. The image will be saved in PNG format.

```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.backend_bases import MouseButton

t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2 * np.pi * t)
fig, ax = plt.subplots()
ax.plot(t, s)


def on_move(event):
    if event.inaxes:
        print(f'data coords {event.xdata} {event.ydata},',
              f'pixel coords {event.x} {event.y}')


def on_click(event):
    if event.button is MouseButton.LEFT:
        print('disconnecting callback')
        plt.disconnect(binding_id)


binding_id = plt.connect('motion_notify_event', on_move)
plt.connect('button_press_event', on_click)

plt.show()
```

- t: Creates an array of time values ranging from 0 to 1, with a step of 0.01.
- s: Computes the sine wave values corresponding to the time values in t.
- fig, ax = plt.subplots(): Creates a figure and axes.
- ax.plot(t, s): Plots the sine wave, where t represents the x-axis (time) and s represents the y-axis (sine wave values).

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

figsrc, axsrc = plt.subplots(figsize=(3.7, 3.7))
figzoom, axzoom = plt.subplots(figsize=(3.7, 3.7))
axsrc.set(xlim=(0, 1), ylim=(0, 1), autoscale_on=False,
          title='Click to zoom')
axzoom.set(xlim=(0.45, 0.55), ylim=(0.4, 0.6), autoscale_on=False,
           title='Zoom window')

x, y, s, c = np.random.rand(4, 200)
s *= 200

axsrc.scatter(x, y, s, c)
axzoom.scatter(x, y, s, c)


def on_press(event):
    if event.button != 1:
        return
    x, y = event.xdata, event.ydata
    axzoom.set_xlim(x - 0.1, x + 0.1)
    axzoom.set_ylim(y - 0.1, y + 0.1)
    figzoom.canvas.draw()

figsrc.canvas.mpl_connect('button_press_event', on_press)
plt.show()


```

- figsrc and axsrc: These are the figure and axes for the source plot, where the user can click to specify the zoom area.
- figzoom and axzoom: These are the figure and axes for the zoomed-in plot that shows a closer view of the region the user clicks on.

- axsrc.set(xlim=(0, 1), ylim=(0, 1)): Sets the x and y limits for the source plot between 0 and 1.
- axzoom.set(xlim=(0.45, 0.55), ylim=(0.4, 0.6)): Sets the initial zoom window for the zoomed-in plot between (0.45, 0.55) on the x-axis and (0.4, 0.6) on the y-axis.
- Both plots have the autoscale_on=False argument, meaning the axes wont automatically rescale when the data changes.

- x, y, s, c = np.random.rand(4, 200): Generates four arrays of random numbers, each containing 200 points.
- x and y represent the coordinates of the scatter points.
- s represents the size of the points.
- c represents the color values.
- s *= 200: Scales the size of the scatter points by multiplying the size values by 200, making the points more visually distinguishable.
- axsrc.scatter(x, y, s, c) and axzoom.scatter(x, y, s, c): Create scatter plots on both the source and zoomed-in axes.

- This function handles mouse click events on the source plot (axsrc).
- event.button != 1: If the event is not triggered by a left mouse button click (i.e., button 1), the function exits without doing anything.
- x, y = event.xdata, event.ydata: Captures the x and y coordinates of the mouse click in data coordinates (i.e., relative to the source plot's scale).
- axzoom.set_xlim(x - 0.1, x + 0.1) and axzoom.set_ylim(y - 0.1, y + 0.1): Adjust the limits of the zoomed-in plot to center around the point where the user clicked. The zoom window is set to span 0.2 units in both the x and y directions (0.1 units before and after the click position).
- figzoom.canvas.draw(): Redraws the zoomed-in figure to reflect the new zoom limits.

This connects the mouse click event ('button_press_event') to the on_press callback function. Every time the user clicks on the source plot, on_press is called, and the zoomed-in plot is updated.
